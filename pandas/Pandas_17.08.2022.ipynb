```python
df6 = df1[['age', 'job', 'marital', 'education']].head()

df6
```

```python
df7 = df2[['age', 'job', 'marital', 'education']].head()
df7
```

```python
# merge
# how - type of the join: 'inner', 'left', 'right', 'outer' | default value is 'inner'
# on - key to merge
pd.merge(df6, df7, on = 'age' )
```

```python
pd.merge(df7, df6, how = 'inner', on = 'age' )
```

```python
pd.merge(df6, df7, how= 'left', on = 'age')
```

```python
pd.merge(df6, df7, how= 'right', on = 'age')
```

```python
pd.merge(df6, df7, how= 'outer', on = ['age', 'job'])
```

```python
# .rename(columns = {'original_column_name':'new_column_name})
df7.rename(columns={'age':'age_1', 'job':'job_1', 'marital': 'marital_1', 'education':'education_1'}, inplace = True )
```

```python
df7
```

```python
# join
df6.join(df7, how = 'outer', on='age')
```

- join() method is used to perform join on row indices and doesn’t support joining on columns unless setting column as index.
- join() by default performs left join.
- merge() method is used to perform join on indices, columns and combination of these two.
- merge() by default performs inner join.
- Both these methods support inner, left, right, outer join types. merge additionally supports the cross join.
